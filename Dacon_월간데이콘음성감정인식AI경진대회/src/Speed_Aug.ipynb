```python
import pandas as pd

train_df = pd.read_csv('./data/train_95_text_aug.csv')
```

```python
from transformers import BertTokenizer, RobertaModel, AutoModel
from transformers import BertModel, RobertaTokenizer, AutoTokenizer
```

```python
import torch
device = torch.device('cuda:1' if torch.cuda.is_available() else 'cpu')
print("Device: {}".format(device))
if device == "cpu":
    raise ValueError("GPU not detected!")
```

```python
train_trans = train_df.transcription
```

```python
import torch
tokenizer = AutoTokenizer.from_pretrained('tae898/emoberta-large')
bert = AutoModel.from_pretrained("tae898/emoberta-large").to(device)
bert.pooler = torch.nn.Identity()
MAX_LEN = 41
```

```python
train_pooled_list = []

for trans in tqdm(train_trans) :
    if type(trans) != str :
            print(trans)
            print(type(trans))
            trans = ''
    item = tokenizer(trans,return_tensors='pt', 
                              max_length=MAX_LEN, 
                              padding='max_length',
                              truncation=True,
                              add_special_tokens=True,
                              return_token_type_ids=True)
    
    input_ids = item['input_ids'][0][None].to(device)

    attention_mask = item['attention_mask'][0][None].to(device)
    _, pooled_output = bert(input_ids= input_ids, attention_mask=attention_mask,return_dict=False)
    pooled_output = pooled_output[0].cpu().detach()
    train_pooled_list.append(pooled_output)
```

```python
import pickle as pk
```

```python
train_dict = {}
for id, pool in zip(train_df.id, train_pooled_list) :
    train_dict[id] = torch.tensor(pool)
    
with open('./data/train_95_pool_aug.pk', 'wb') as f:
    pk.dump(train_dict, f)
```
