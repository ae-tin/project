```python
print("문자열 안에 "가 포함")
```

### 여러 줄인 문자열을 변수에 대입하고 싶을 때

한번에 여러 줄인 문자열을 사용하려고 할 때는  
''' 또는 """ 로 묶어준다.


```python
multiline='''
오늘은 2021년 7월 27일이다.
겁나 덥다.
배도 고프다.
'''
```

```python
print(multiline)
```

### 문자열에 작은 따옴표와 큰 따옴표 모두 포함될 때(이스케이프 문자 사용)

\n 문자열 안에서 줄을 바꿈  
\t 문자열 안에서 탭 간격을 줌  
\\ 문자열 안에 \를 출력하고 싶을 때 사용  
\' 문자열 안에 '를 출력하고 싶을 때 사용  
\" 문자열 안에 "를 출력하고 싶을 때 사용  

```python
str1 = "문자열 안에서 \n줄 바꿈"
print(str1)
```

```python
str2 = "문자열 안에서 \t 탭으로 간격을 맞춤"
print(str2)
```

```python
str3 = "문자열에서 \\ 를 출력"
print(str3)
```

```python
str4 = "문자열에서 \' 를 출력"
print(str4)
```

```python
str5 = "문자열에서 \" 를 출력"
print(str5)
```
