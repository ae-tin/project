### 파이썬 내부 모듈 - 

```python
import requests

req = requests.get('https://www.naver.com/')
html = req.text

print(html)
```

모두 문자열로 저장돼있기 때문에 노가다 인덱스 하게 될 수도 있으니 다른 걸 이용해 준다.  
BeautifulSoup in bs4!!1

```python
import requests
from bs4 import BeautifulSoup

req = requests.get('https://www.naver.com/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')

my_classes = soup.select('strong.title')

for my_class in my_classes :
    print(my_class.text)
```

그래도 원하는 걸 찾기엔 너무 많이 나옴. 그래서 <strong > 안에 class= title을 찾아준다
    strong.title 파이썬처럼 속성.클래스명으로 찾음

```python
import requests
from bs4 import BeautifulSoup

req = requests.get('https://datalab.naver.com/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')

my_classes = soup.select('span.title')

for my_class in my_classes :
    print(my_class.text)
```

strong.title은 class가 title인것

### 밑에는 구글 드라이버를 통해서 주소로 접속하는 모듈

### find 사용법

find_element_by_id('id)  
-> id는 1개 밖에 없어서 element 가 맞음
find_element_by_name('name')  
find_element_by_xpath('xpath')  
find_element_by_tag_name('태그이름)    >> span, input, body 같은 태그들  
find_element_by_class_name('클래스')  
-> 여러개 찾고 싶을 땐 리스트 형식으로 저장되는 elements 로 검색

```python
from selenium import webdriver

driver = webdriver.Chrome('c:/test/chromedriver.exe')

driver.get('https://top.cafe.daum.net/')

results = driver.find_elements_by_class_name('desc_info')
#elements 는 리스트로 저장됨

for result in results :
    print(result.text)
```
