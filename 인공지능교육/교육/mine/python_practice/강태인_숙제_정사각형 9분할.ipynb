## 숙제

어떤 게임에 N개의 스테이지(스테이지수 랜덤한 한자리수)가 있다.

이 각 스테이지에는 처치해야할 몬스터들이 있다.

플레이어는 처음 출발할 때 총알이 없어서 각 스테이지의 상점에서 총알을 구입하고 출발하여야 한다.

탄창의 크기는 무제한이어서 얼마든지 많은 총알을 살 수 있다.

몬스터를 처치할 때마다 하나의 총알을 사용한다.

각 스테이지에는 단 하나의 상점이 있으며, 스테이지 마다 상점의 총알 가격은 다를 수 있다.

예를 들어, 다음처럼 3개의 스테이지가 있다고 하자.

총알 7원(몬스터3) - 총알 3원(몬스터4) - 총알 5원(몬스터2)
제일 왼쪽 스테이지에서 총알 3 + 4 + 2 개를 사서 최종 스테이지까지 클리어하면

(3 + 4 + 2)*7 = 63원이다.

만약 제일 왼쪽 스테이지에서 3개의 총알을 사고(3×7 = 21원) 다음 번 스테이지에서 총알 4개를 사고(3×4 = 12원) 다음 스테이지에서 총알 2개를 사면(5×2 = 10원) 43원이다.

각 스테이지의 총알의 가격과 몬스터의 수 및 스테이지 수를 랜덤으로 생성(한자리수)하고 왼쪽부터 오른쪽으로 차례로 스테이지를 모두 클리어할 때 가장 최소한의 금액으로 모든 스테이지를 클리어하는 금액을 구하는 프로그램


---------------------------------------------------

일단 최소 금액으로 스테이지를 모두 클리어 해야하는데, 변수로 생각해야 할 것은 스테이지 수, 총알 가격, 몬스터의 수 이다.  
  
변수들 중 금액에 직접적인 영향을 주는 것은 총알 가격과 몬스터의 수 이다.  
  
  그리고 최소의 금액을 구하려면 먼저 __몬스터의 수 = 필요한 총알의 수__ 이므로 몬스터의 수를 초과한 총알은 불필요 하다.  
  
  또 생각해야 할 것은 각 스테이지의 총알의 가격인데 모든 스테이지의 총알 가격을 알 수 있으니  
  
  먼저 __가장 낮은 총알 가격의 스테이지 순서를 파악__ 하고 그 스테이지에서는 뒤의 스테이지 것까지 다 사야한다.  
  
  만약, 가장 낮은 총알 가격의 스테이지의 순서가 첫 번째가 아니라면 두 번째로 낮은 가격의 스테이지를 파악하고,  
  
  또, 두 번째로 낮은 가격의 스테이지가 첫 번째가 아니라면 세 번째로 낮은 가격의 스테이지를 파악해야 할 것이다. 
  
  반복...
  
  왜냐하면 스테이지는 클리어 하고 가야하기때문에 첫 번째 스테이지에서는 낮은 가격의 스테이지가 나오기 전까지 무조건 총알을 구입해야하고  
  
  낮은 가격의 스테이지가 나오면 최소 금액의 스테이지가 나오기 전까지의 총알을 사야하기 때문이다.
  
  그리고 각 스테이지의 몬스터수와 필요한 총알 가격을 곱해주면 최소금액 끝!
  
  

클래스로 만들면,  

    속성에
        스테이지 수 (int)
        총 몬스터 수 (int)
        스테이지마다 몬스터 수 (리스트)
        스테이지마다 총알 가격 (리스트)
    
    기능
        초기설정 
            스테이지수 랜덤하게 만들기
            스테이지수 따라 랜덤한 몬스터 수 리스트에 넣기
            총 몬스터수 구하기
            스테이지수 따라 랜덤한 총알가격 리스트에 넣기
        
        가장 낮은 총알 가격 순서 파악 
            
        위의 기능으로 가장 낮은 총알 가격 순서 앞에 다음으로 낮은 총알 가격 있는지 0번째까지 알아보고 리스트로 순서 출력 
        
        위의 기능으로 스테이지 별 총알 개수 정하기
        
        최소 금액 구하기
        





```python
from random import *
class Game :
    monster = []
    bullet = []
    stage = 0
    m_num = 0
    min1 = 10
    idx = 10
    idx_min = []
    def __init__(self) :
        self.monster = []
        self.bullet = []
        self.stage = randint(1,9)
        m_num = 0
        min1 = 10
        idx = 10
        idx_min = []
        for i in range(0,self.stage,1) :
            num = randint(1,9)
            self.monster.append(num)
            self.bullet.append(randint(1,9))
            self.m_num = self.m_num + num
    
    def bullet_min(self,stage) :
        min1 = 10
        idx = 10
        for i in range(0,stage,1) :
            if self.bullet[i] < min1 :
                min1 = self.bullet[i]
                idx = i
        return idx
    def bullet_search(self) :
        a = self.stage
        idx_min = []
        while a != 0 :
            a = self.bullet_min(a)
            idx_min.append(a)
        return idx_min
    
    # 여기까지 일단 만들었습니다~~
    
    def minimize(self) :
        result = 0
        price = self.bullet[0]
        for i in range(0,a.stage,1) :
            if price >self.bullet[i] :
                price = self.bullet[i]
            result = result + self.monster[i] * price
        return result
            
        
    
    
    
    
```

## 문제

플레이어들은 하나의 아이템만 소유 할 수 있다.  
  
  아이템이 N종류 있다. N은 랜덤한 수로 생성(1-9)

  내가 가지고 있는 아이템은 1번이다.  
  
  같은 아이템을 가지고 있는 플레이어의 수가 제일 많은 플레이어들이 승리  
  
  1번 2번 3번 아이템이 있을 떄

1번 아이템을 가진 플레이어 3명  
2번 아이템을 가진 플레이어 4명  
3번 아이템을 가진 플레이어 4명

그래서 2번 아이템을 가진 플레이어와 3번 아이템을 가진 플레이어들을  
1번 아이템으로 아이템을 변경하도록 매수  
  
  
1번 아이템을 가진 플레이어 5명  
2번 아이템을 가진 플레이어 3명  
3번 아이템을 가진 플레이어 3명  
  
  
내가 매수해야하는 플레이어의 최솟값을 구하는 프로그램  


```python
from random import *
class Game2 :
    item = 0
    player_item = [1]
    player_num = 0
    count = 0
    def __init__(self) :
        self.item = randint(1,9)
        self.player_item = [1]
        self.player_num = randint(1,9)
        count = 0
        for i in range(1,self.player_num,1) :
            num = randint(1,self.item)
            self.player_item.append(num)
    def compare(self) :
        cnt = 0
        d = []
    
        for i in range(1,self.item+1,1):
            cnt = 0
            for j in self.player_item :
                if i == j :
                    cnt = cnt + 1
            d.append(cnt)
        
        return d
    
    def count(self) :
        com = self.compare()[0]
        for i in range(1,len(self.compare()),1) :
            if com <= self.compare()[i] :
                return '이대로는 지고 말겠어'
                break
            else :
                return '이겼다'
   
                    
                    
                    
                    
                    
                    
                    
```

  G : 현재 위치에서 출발 지점까지의 총 cost 합

  이동한 곳에서 주변 좌표들의 f g h 값을 구하고 open 리스트에 넣는다.  
  
  현재 위치가 목표지점이 되면 반복문 종료
