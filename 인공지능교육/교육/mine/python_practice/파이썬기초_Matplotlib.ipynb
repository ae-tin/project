# Matplotlib

import matplotlib.pyplot as plt  
  
plt.plot(x축,y축, format, parameter) : 그래프의 내용에 대한 데이터  
plt.xlabel('x축이름') : x축이름 설정  
plt.ylabel('y축이름') : y축이름 설정  
*폰트변경 : fontdict로 폰트명, 폰트사이즈, 폰트굵기, 폰트 칼라 등을 지정


```python
import matplotlib.pyplot as plt

data_plot = {'xdata': [1,2,3,4], 'ydata':[2,9,11,15]}

plt.plot('xdata','ydata', data = data_plot, color = 'red')
plt.show()
```

```python
data_plot1 = {'xdata': [1,2,3,4], 'ydata':[2,9,11,15,18]}  # x,y 크기 다르게 지정

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue')
```

x,y의 크기(shape)가 같아야 일대일 대응으로 그릴 수 있다. 


```python
data_plot1 = {'xdata': [1,4,3,2], 'ydata':[2,9,11,15,]}  # x,y 크기 다르게 지정

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue')
```

x가 정렬 안돼있으면 정렬안된 순서따라 감

```python
data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]}  # x,y 크기 다르게 지정

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue')
plt.xlabel('xlab',labelpad = 50)  # 축과 간격 떨어뜨리기
plt.ylabel('ylab',labelpad = 10)  # labelpad = 간격
plt.show()
```

```python
plt.plot([1,2,3,4,5,6,7], [13,2,34,56,8,10,25], color= 'blue')
plt.xlabel('xlab',labelpad = 20)  # 축과 간격 떨어뜨리기
plt.ylabel('ylab',labelpad = 10)  # labelpad = 간격
plt.show()
```

```python
plt.plot([1,2,3,4,5,6,7], [13,2,34,56,8,10,25], color= 'blue')
plt.xlabel('xlab',labelpad = 20,fontdict = {'family':'serif','color':'red',
                                           'weight' : 'normal','size':20})  # 축과 간격 떨어뜨리기
plt.ylabel('ylab',labelpad = 10)  # labelpad = 간격
plt.show()
```

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue')
plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.show()
```

loc같은 경우에 x축은 right, left, y축은 top, bottom으로 위치를 지정 할 수 있다.

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 
data_plot2 = {'xdata': [1,2,3,4,5], 'ydata':[6,2,1,20,12]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue')
plt.plot('xdata', 'ydata', data = data_plot2, color= 'yellow')

plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.legend()
plt.show()
```

plt.legend()를 하면 범례가 들어가지만 plot에서 label설정을 해줘야 설정대로 들어감  
legend에 loc으로 위치 설정 해줄 수 있음. lower right, right, center, best, 좌표설정도 가능함

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 
data_plot2 = {'xdata': [1,2,3,4,5], 'ydata':[6,2,1,20,12]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue',label = 'btc')
plt.plot('xdata', 'ydata', data = data_plot2, color= 'yellow',label = 'eth')

plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.legend(loc = (1.0,0.1))
plt.show()
```

__그래프 축 지정__  
  
  - 정확한 범위 표현을 취해 최소값, 최대값을 표시한것. x,y,별로 지정
  - plt.xlim([xmin,xmax]) : x축의 최소, 최대 범위지정
  - plt.ylim([ymin,ymax]) : y축의 최소, 최대 범위지정
    
  - plt.axis([xmin,xmax,ymin,ymax]) : x,y축 동시 지정,
  - axis 추가 기능
  - off, on square, scaled, tight

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 
data_plot2 = {'xdata': [1,2,3,4,5], 'ydata':[6,2,1,20,12]} 
data_plot3 = {'xdata': [1,2,3,4,5], 'ydata':[4,5,6,7,8]} 
data_plot4 = {'xdata': [1,2,3,4,5], 'ydata':[16,21,11,5,1]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue',label = 'btc')
plt.plot('xdata', 'ydata', data = data_plot2, color= 'yellow',label = 'eth')
plt.plot('xdata', 'ydata', data = data_plot3, color= 'orange',label = 'crd')
plt.plot('xdata', 'ydata', data = data_plot4, color= 'deeppink',label = 'bcc')

plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.legend(loc = (1.0,0.1),fontsize = 20)
plt.axis()
plt.show()
```

__그래프 선 바꾸기__  
  
  - plt.plot(x축,y축, 선종류) : 그래프에 표시되는 선 종류 선택, 선모양 직접입력, 
                                  linestyle 사용
  - ------ : dashed  '--'  
  - ...... : dotted  ':'  
  - ._._.__._ : dash-dot '-.'  

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 
data_plot2 = {'xdata': [1,2,3,4,5], 'ydata':[6,2,1,20,12]} 
data_plot3 = {'xdata': [1,2,3,4,5], 'ydata':[4,5,6,7,8]} 
data_plot4 = {'xdata': [1,2,3,4,5], 'ydata':[16,21,11,5,1]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue',
         label = 'btc', linestyle = 'dashed')   
plt.plot('xdata', 'ydata',':', data = data_plot2, color= 'red', 
         label = 'eth')  # linestyle로 지정할거 아니면 무조건 x,y 뒤에 나와야함
plt.plot('xdata', 'ydata','-.', data = data_plot3, color= 'orange',
         label = 'crd')
plt.plot('xdata', 'ydata', data = data_plot4, color= 'deeppink',
         label = 'bcc')

plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.legend(loc = (1.0,0.1),fontsize = 20)
plt.axis()
plt.show()
```

__마커 찍기__  - 데이터 값에 점 찍기.
  
  - 위에서 선종류 선택할 때에 '--' 선종류와 같이 'o'를 넣어줌
  - 'color,[형태,[선모양]]' 순으로 적어준다
  - 또는 marker = 'x','o','d','s'등 넣을 수 있음

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 
data_plot2 = {'xdata': [1,2,3,4,5], 'ydata':[6,2,1,20,12]} 
data_plot3 = {'xdata': [1,2,3,4,5], 'ydata':[4,5,6,7,8]} 
data_plot4 = {'xdata': [1,2,3,4,5], 'ydata':[16,21,11,5,1]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue',
         label = 'btc', linestyle = 'dashed')   
plt.plot('xdata', 'ydata',':', data = data_plot2, color= 'deeppink', 
         label = 'eth')  # linestyle로 지정할거 아니면 무조건 x,y 뒤에 나와야함
plt.plot('xdata', 'ydata','o-.', data = data_plot3, color= 'orange',
         label = 'crd')
plt.plot('xdata', 'ydata', 'rs-',data = data_plot4, label = 'bcc')

plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.legend(loc = (1.0,0.1),fontsize = 20)
plt.axis()
plt.show()
```

__그래프 영역 색깔 채우기__  
  
  - 그래프 상 특정 위치에 색을 채워 강조함  
  - plt.fill_between(data[x축범위],data[y축범위], alpha = value)
  - 여기서 x축 범위의 크기와 y축 범위의 크기가 같아야한다
  
- 두개의 그래프 사이를 채울수도 있다

```python
font1 = {'family':'serif','color':'deeppink','weight' : 'normal','size':20}

data_plot1 = {'xdata': [1,2,3,4,5], 'ydata':[2,9,11,15,18]} 
data_plot2 = {'xdata': [1,2,3,4,5], 'ydata':[6,2,1,20,12]} 
data_plot3 = {'xdata': [1,2,3,4,5], 'ydata':[4,5,6,7,8]} 
data_plot4 = {'xdata': [1,2,3,4,5], 'ydata':[16,21,11,5,1]} 

plt.plot('xdata', 'ydata', data = data_plot1, color= 'blue',
         label = 'btc', linestyle = 'dashed')   
plt.plot('xdata', 'ydata',':', data = data_plot2, color= 'deeppink', 
         label = 'eth')  # linestyle로 지정할거 아니면 무조건 x,y 뒤에 나와야함
plt.plot('xdata', 'ydata','o-.', data = data_plot3, color= 'orange',
         label = 'crd')
plt.plot('xdata', 'ydata', 'rs-',data = data_plot4, label = 'bcc')

plt.xlabel('xlab',labelpad = 10,fontdict = font1,loc = 'right') 
plt.ylabel('ylab',labelpad = 50, loc = 'top') 
plt.legend(loc = (1.0,0.1),fontsize = 20)
plt.axis([0,6,0,30])
plt.fill_between(data_plot2['xdata'][1:4],
                 data_plot2['ydata'][1:4],
                 data_plot3['ydata'][1:4],alpha = 0.7)
plt.show()
```
