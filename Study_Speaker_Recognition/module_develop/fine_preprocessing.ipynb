### veri_test_ver1 
- 같은 speaker면 1, 다른 speaker면 0(spoof의 speaker라도 speaker로 보고 검증을 한다)

```python
with open('/Data1/VoxCeleb/ASVspoof2019/LA/ASVspoof2019_LA_train/dev_list_ver1.txt', "r") as v:
    v_meta = v.readlines()
    
speaker = []
for line in v_meta:
    spk, fnm = line.strip().split(" ")
    speaker.append(spk)
speaker = list(set(speaker))
print(speaker)
spk_dic = {i:[] for i in speaker}
print(spk_dic)
```

```python
for line in v_meta:
    spk, fnm = line.strip().split(" ")
    spk_dic[spk].append(fnm)
```

```python
import random

veri_test_ver1 = []
for idx, spk in enumerate(speaker) :
    d = list(range(0,len(speaker)))                                  # 다른 spk 뽑기 위한 speaker의 인덱스 리스트
    d.pop(idx)                                                       # 현재 spk의 인덱스 제거
    e = random.randint(0,len(speaker) - 2)                           # spk인덱스가 없으므로 d의 인덱스 난수 e 생성
    for idx1, fnm in enumerate(spk_dic[spk]) :
        a = 0
        b = list(range(0,len(spk_dic[spk])))                # b는 현재 spk의 utt 인덱스 리스트 
        f = list(range(0,len(spk_dic[speaker[d[e]]])))               # 다른 spk의 utt 인덱스 리스트
        b.pop(idx1)                                          # b에서 현재 spk의 utt의 인덱스를 제외
        f.pop(random.randint(0,len(spk_dic[speaker[d[e]]])-1))       # f에서 다른 spk의 utt 인덱스를 랜덤으로 제외
        while a < 8 :
            c = random.randint(0, len(spk_dic[spk]) - 2)    #b에서 현재 spk의 utt 인덱스가 없으므로 b의 인덱스 난수 c 생성
            g = random.randint(0, len(f)-1)                            # f의 인덱스(다른 spk의 utt인덱스)중 랜덤
            if a % 2 == 0 :
                test = '1 ' + fnm + ' ' + spk_dic[spk][b[c]] #같은 spk중에 다른 utt
                veri_test_ver1.append(test)
            else : 
                test = '0 ' + fnm + ' ' + spk_dic[speaker[d[e]]][f[g]]  # 다른 spk의 랜덤 utt
                veri_test_ver1.append(test)
            a = a+1
                
            
```
